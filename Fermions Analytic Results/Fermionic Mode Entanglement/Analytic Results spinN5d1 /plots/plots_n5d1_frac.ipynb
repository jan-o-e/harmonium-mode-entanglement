```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import gridspec, rc
import pandas as pd
```

```python
epmode0neg= epdata[0]
epmode1neg = epdata[1]
epmode2neg = epdata[2]
epmode3neg =epdata[3]
epmode4neg = epdata[4]
epmode6neg = epdata[5]
epmode8neg = epdata[6]

epmode0pos= epdata[7]
epmode1pos = epdata[8]
epmode2pos = epdata[9]
epmode3pos =epdata[10]
epmode4pos = epdata[11]
epmode6pos = epdata[12]
epmode8pos = epdata[13]

enmode0neg= endata[0]
enmode1neg = endata[1]
enmode2neg = endata[2]
enmode3neg =endata[3]
enmode4neg = endata[4]
enmode6neg = endata[5]
enmode8neg = endata[6]

enmode0pos= endata[7]
enmode1pos = endata[8]
enmode2pos = endata[9]
enmode3pos =endata[10]
enmode4pos = endata[11]
enmode6pos = endata[12]
enmode8pos = endata[13]



kappalist_neg = np.linspace(-0.99,-0.01, num=len(epmode0neg))
kappalist_pos = np.linspace(0.01, 5, num=len(epmode0pos))

kappalist_pos_log=[]

for x in kappalist_pos:
    kappalist_pos_log.append(10**(x))

rc('font', **{'family': 'serif', 'serif': ['Computer Modern']})
rc('text', usetex=True)

fig = plt.figure(figsize=(10,8))
gs = fig.add_gridspec(2, 2, hspace=0, wspace=0,)
(ax1, ax2), (ax3, ax4) = gs.subplots(sharex='col', sharey='row')
ax1.plot(kappalist_neg, epmode0neg, label='$m_0$', color='black', linewidth=2.5)
ax1.plot(kappalist_neg, epmode1neg, label='$m_1$', color='grey', linewidth=2.5)
ax1.plot(kappalist_neg, epmode2neg, label='$m_2$', color='black', linestyle='dashed', linewidth=2.5)
ax1.plot(kappalist_neg, epmode3neg, label='$m_3$', color='grey', linestyle='dashed', linewidth=2.5)
ax1.plot(kappalist_neg, epmode4neg, label='$m_4$', color='black', linestyle='dotted', linewidth=2.5)
ax1.plot(kappalist_neg, epmode6neg, label='$m_6$', color='grey', linestyle='dotted', linewidth=2.5)
ax1.tick_params(axis='y', width=1, labelsize=20)
ax1.legend(loc = 'best', fontsize=18, ncol=2)
ax1.set_ylabel('$E^{P}/E$', fontsize=22)
ax3.set_xlabel('Coupling strength / $\kappa$', fontsize=22)
ax3.set_ylabel('$E^{N}/E$', fontsize=22)
ax4.set_xlabel('Coupling strength / $\kappa$', fontsize=22)
ax2.plot(kappalist_pos, epmode0pos, label='$m_0$', color='black', linewidth=2.5)
ax2.plot(kappalist_pos, epmode1pos, label='$m_1$', color='grey', linewidth=2.5)
ax2.plot(kappalist_pos, epmode2pos, label='$m_2$', color='black', linestyle='dashed', linewidth=2.5)
ax2.plot(kappalist_pos, epmode3pos, label='$m_3$', color='grey', linestyle='dashed', linewidth=2.5)
ax2.plot(kappalist_pos, epmode4pos, label='$m_4$', color='black', linestyle='dotted', linewidth=2.5)
ax2.plot(kappalist_pos, epmode6pos, label='$m_6$', color='grey', linestyle='dotted', linewidth=2.5)
ax2.set_xticks([1,2,3,4,5], labels=['$1$', '$10$', '$10^2$', '$10^3$', '$10^4$'])
ax2.tick_params(axis='x', width=1, labelsize=20)
ax3.tick_params(axis='y', width=1, labelsize=20)


ax3.plot(kappalist_neg, enmode0neg, label='$m_0$', color='black', linewidth=2.5)
ax3.plot(kappalist_neg, enmode1neg, label='$m_1$', color='grey', linewidth=2.5)
ax3.plot(kappalist_neg, enmode2neg, label='$m_2$', color='black', linestyle='dashed', linewidth=2.5)
ax3.plot(kappalist_neg, enmode3neg, label='$m_3$', color='grey', linestyle='dashed', linewidth=2.5)
ax3.plot(kappalist_neg, enmode4neg, label='$m_4$', color='black', linestyle='dotted', linewidth=2.5)
ax3.plot(kappalist_neg, enmode6neg, label='$m_6$', color='grey', linestyle='dotted', linewidth=2.5)
ax3.set_xticks([-1,-0.75,-0.5,-0.25,0])
ax3.tick_params(axis='x', width=1, labelsize=20)
ax4.plot(kappalist_pos, enmode0pos, label='$m_0$', color='black', linewidth=2.5)
ax4.plot(kappalist_pos, enmode1pos, label='$m_1$', color='grey', linewidth=2.5)
ax4.plot(kappalist_pos, enmode2pos, label='$m_2$', color='black', linestyle='dashed', linewidth=2.5)
ax4.plot(kappalist_pos, enmode3pos, label='$m_3$', color='grey', linestyle='dashed', linewidth=2.5)
ax4.plot(kappalist_pos, enmode4pos, label='$m_4$', color='black', linestyle='dotted', linewidth=2.5)
ax4.plot(kappalist_pos, epmode6pos, label='$m_6$', color='grey', linestyle='dotted', linewidth=2.5)
ax4.tick_params(axis='x', width=1, labelsize=20)

fig.savefig('ferm_spin_n5d1_efrac.pdf')
```
